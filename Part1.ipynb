### Разделение обращений и ответов
В `payload.body` хранится информация об обращениях граждан, а также об ответах на запросы. При просмотре данных, можно увидеть, что обращения содержатся в виде списка словарей, а обращение в виде обычной строки

При таком раскладе создадим переменные, содержащие строки как тип `str`, так и `list`. При последующем объединении в `DataFrame` необходимо наличие `sidebar.id` обращения, чтобы образовать соединение таких данных

```python
df_str=df[df['payload.body'].map(lambda x: type(x)==str)]
```

```python
df_str.head(2)
```

```python
column_list=pd.DataFrame(df[df['payload.body'].map(lambda x: type(x)==list)][['payload.body', 'sidebar.id']])
```

```python
column_list.head(2)
```

Чтобы произвести объединение, воспользуемся командой `merge`. Как и было указано, воспользуемся `sidebar.id`, чтобы чтобы объединить эти данные. И укажем постфиксы как ответ(answer) и обращение(appeal)

```python
df=df_str.merge(column_list, on=('sidebar.id'), suffixes=('_answer', '_appeal')).reset_index(drop=True)
```

```python
df.head(4)
```

На выходе получаем атрибуты `payload.body_answer` и `payload.body_appeal`

### 1.2. Создание структуры данных для итогового набора полей

Создадим новый атрибут `appeal`, который содержит данные обращения, которые находятся в `payload.body_appeal`

Создание этого атрибута необходимо, чтобы получить данные текста, которые в будущем можно будет подать в алгоритмы кластеризации или классификации

```python
df['payload.body_appeal'][0]
```

Как видно выше, у нас есть список словарей, где данные обращения находятся в ключе `text`. Соберём все данные ключей `text` для каждого поля в наборе данных. Ничего страшного, если это будет выглядеть "криво", главное - это получение данных текста обращения, а потом уже можно сделать преодобработку

```python
#Функция получения ключей text в списке словарей
def get_text(df):
    str_=''
    for dict_ in (x for x in df if x['text']):
        str_+=dict_['text'] + ' '
    return str_
```

```python
#Применение функции получения ключей text для каждого поля в новый атрибут appeal
df['appeal']=df['payload.body_appeal'].apply(get_text)
df.drop('payload.body_appeal', axis=1, inplace=True)
```
